```python
years = df.columns[1:]  

total_values = df[years].apply(pd.to_numeric, errors='coerce').sum()

year_labels = [str(year) for year in years]

plt.figure(figsize=(30, 30))  
plt.barh(year_labels, total_values, color='#191970')
plt.xlabel('Total Values')
plt.ylabel('Year', size=20)
plt.title('Total Values per Year', size=20)
plt.show()
```

```python
country_by_1960= df.sort_values(by='1960').head(20)
country_by_1960
```

```python
country_by_1960_t = country_by_1960.set_index('Country Name').T

country_by_1960_t = country_by_1960_t.apply(pd.to_numeric, errors='coerce')

for country_name in country_by_1960_t.columns:
    y_values = country_by_1960_t[country_name].dropna()


    if y_values.empty:
        continue

    plt.figure(figsize=(12, 5))
    sns.barplot(x=y_values.index.astype(str), y=y_values.values)
    plt.title(f"{country_name} - Data Values from 1960 to 2022")
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

```

```python
country_by_2022 = df[['Country Name', '2022']].sort_values(by='2022', ascending=False).head(20)

country_by_2022_t = country_by_2022.set_index('Country Name').T


country_by_2022_t = country_by_2022_t.apply(pd.to_numeric, errors='coerce')


for year_name, data_values in country_by_2022_t.iterrows():
    data_values = data_values.dropna() 

    if data_values.empty:
        continue 

    fig = plt.figure(figsize=(10, 5))
    sns.barplot(x=data_values.index, y=data_values.values)
    plt.xlabel('Country')
    plt.ylabel('Data Value')
    plt.title(f"{year_name} - Data Values per Country")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
